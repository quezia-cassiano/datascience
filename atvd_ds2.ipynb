```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
paises = ['Estados Unidos', 'China', 'Japão',
            'Alemanha', 'Reino Unido',
            'França', 'Índia',
            'Itália', 'Brasil',
            'Canadá']
```

```python
dolares = [18.6, 11.2, 4.9,
           3.4, 2.6, 2.5,
           2.2, 1.8, 1.8,
           1.5]
```

```python
plt.rcParams["figure.figsize"] = (15,8) #redimensionar
plt.bar(paises, dolares, color='#ccd5ae')
plt.title('Países com maiores PIB em 2023', fontsize=22)
plt.legend(['Trilhões de doláres'], loc = 'right', fontsize=14)
plt.xlabel('Países', fontsize=16)
plt.ylabel('Doláres', fontsize=16)
plt.show() #so pra exibir a figura
```
