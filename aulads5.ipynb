```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-talk')
```

```python
dias = [1,2,3,4,5]
dormir = [7,8,6,77,7]
comer = [2,3,4,5,3]
trabalhar = [7,8,7,2,2]
passear = [8,5,7,8,13]
```

```python
plt.stackplot(dias, dormir, comer, trabalhar, passear,
              labels=['dias', 'dormir', 'comer', 'trabalhar', 'passear'],
              colors = ['m', 'c', 'r', 'k', 'b'])
plt.legend(loc='upper left')
plt.show()
```

```python
porcentagens  = [70,20,20,130]
cidades = ['SÃ£o Paulo', 'Recife', 'Fortaleza', 'Manaus']
colunas = ['c', 'm', 'r', 'k']
```

```python
plt.pie(porcentagens, labels = cidades, colors = colunas,
        startangle = 90, shadow = True, explode = (0, 0.15, 0, 0))
#plt.savefig('pie_chart.png', dpi=600, size(100, 100))
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax1.plot(np.random.randn(50), color ='red')

ax2 = fig.add_subplot(1,2,2)
ax2.scatter(np.arange(50), np.random.randn(50))
plt.show
```

```python
x = np.linspace(0, 1., 100)
fig, axes = plt.subplots(1, 2, figsize = (12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("Eixos com range padrÃ£o")

x = np.linspace(-1, 1., 100)
axes[1].plot(x, x**2, x, x**3)
axes[1].set_ylim([0, 1])            # Especifico o tamanho do eixo y
axes[1].set_xlim([-1, 1])           # Especifico o tamanho do eixo x
axes[1].set_title("Eixos customizados");
```
